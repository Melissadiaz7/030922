# Visualizaci√≥n de Datos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Scatter plot (diagramas de dispersion)
x = np.arange(1,20)
y = x**2
y1 = x**2 + 5*x

plt.scatter(x,y, s=40, color='red', marker='s', label='x^2')
plt.scatter(x,y1, s=40, color='blue', marker='s', label='x^2 + 5x')
plt.legend(title='Funciones')
plt.title('Tipos de Funciones', fontsize=20)
plt.xlabel('x', color='green')
plt.ylabel('f(y)')
plt.grid()

#plt.xlim(0,5)

plt.text(5, 300, 'este texto', color='brown')
 
 # Agregar las coordenadas de los marcadores en texto

for val_x in x:
    plt.text(val_x, val_x**2,f'({val_x},{val_x**2}')
```

```python
# Histogramas

media = 20
std = 3
numeros = np.random.normal(media, std, 1000)

plt.hist(numeros, bins=20, cumulative=True, density=True);   #; para no imprimir, acumular los valores
plt.title('Distribucion Normal')
plt.xlabel('Numeros')
plt.ylabel('Frecuencia')
```

```python
plt.boxplot(numeros, vert=False);
```
